```python
copy_df = spotify_df.copy()
```

```python
def contains_pop(genre):
    word = "pop"
    if word in genre:
        return word
    else:
        return genre

copy_df['top genre'] = copy_df['top genre'].map(lambda x: contains_pop(x))
copy_df['top genre'].value_counts()
```
